```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")

# Preview data
df.head()
```

```python
df.info()

df.isnull().sum()
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df['duration'] = df['duration'].fillna("Unknown")
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['month_added'] = df['date_added'].dt.month
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration_minutes'] = movies['duration'].str.replace(' min','')
movies['duration_minutes'] = pd.to_numeric(movies['duration_minutes'], errors='coerce')
```

```python
shows = df[df['type'] == 'TV Show'].copy()

shows['seasons'] = shows['duration'].str.replace(' Seasons','')
shows['seasons'] = shows['seasons'].str.replace(' Season','')
shows['seasons'] = pd.to_numeric(shows['seasons'], errors='coerce')
```

```python
genres = df.copy()

genres['listed_in'] = genres['listed_in'].str.split(', ')

genres = genres.explode('listed_in')
```

```python
countries = df.copy()

countries['country'] = countries['country'].str.split(', ')

countries = countries.explode('country')
```

```python
df.to_csv("netflix_cleaned_main.csv", index=False)

genres.to_csv("netflix_genres.csv", index=False)

countries.to_csv("netflix_countries.csv", index=False)

movies.to_csv("netflix_movies_duration.csv", index=False)

shows.to_csv("netflix_tvshows_seasons.csv", index=False)
```
